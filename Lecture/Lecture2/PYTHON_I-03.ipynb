```python
"""Przykład PY11. Ciąg Fibonacciego F01."""
def fib01(n):    # wypisz ciąg Fibonacciego az do n
   "Wypisuje ciąg Fibonacciego aż do n"
   a, b = 0, 1
   while b < n:
       print(b,)
       a, b = b, a+b

fib01(2000)         
```

```python
"""Przykład PY12. Ciąg Fibonacciego F02."""
def fib02(**v):    # wypisz ciąg Fibonacciego az do n
   "Tworzy ciąg Fibonacciego aż do n"
   a,b,wyn = 0,1,[]
   while b < v['n']:
       wyn.append(b)
       a,b=b,a+b
   return wyn

v={'n':2000, 'a':20, 'x':'abc'}
print(fib02(**v))
```

```python
"""Przykład PY13. Przykład generowania uogólnionego ciągu Fibonacciego."""
def fib03(n,p=2, err='Error'): #wypisz uogolniony cg Fibonacciego do n
  "Tworzy ciąg Fibonacciego aż do n a_n=a_{n-p-1}+a_{n-p}+...+a_{n-1}"
  if p>1:
    a = [0 for i in range(0,p-1)]
    a.append(1)
    b=sum(a)
    wyn = []
    while b < n:
      a=a[1:]
      a.append(b)
      wyn.append(b)
      b=sum(a)
    return wyn
  else:
    return err
print(fib03(1000))
print(fib03(1000,3))
print(fib03(1000,4))
```

```python
""" Przykład PY17. Rekurencja Flawiusza """
import math
def flawiusz(n):
    l = n - (1 << (n.bit_length() - 1))    
    return 2*l+1

def flawiusz1(n,k):
    D=1
    while D<=n:
        D=math.ceil(k*D/(k-1))
    return k*n+1-D

print(flawiusz(145))
print(flawiusz1(145,2))
print(flawiusz1(k=4, n=121))
```

```python
""" Algorytm Euklidesa"""
def NWD(n,m):
    return (n if m==0 else NWD(m,n % m))

def Euklid(n,m):
    while m != 0:        
        n, m = m, n % m
    return n

def EuklidR(n,m):
    n0,m0=n,m
    np,mp=1,0
    r,s=0,1
    while (m0!=0):
        iloraz,reszta=divmod(n0,m0)    
        n0,m0=m0,reszta
        newr,news=np-iloraz*r, mp-iloraz*s
        np,mp=r,s
        r,s=newr,news        
    return [np,mp,n0]

print(NWD(165,44))
print(Euklid(165,44))
print(EuklidR(165,44))
```

```python
"""Przykład PY19. Metoda Newtona znajdowania zera funkcji."""
def pochodna(f):
    def oblicz(x):
        return (f(x+dx)-f(x))/dx
    return oblicz
def metoda_newtona(f,x,dx,eps):
    df=pochodna(f)
    while 1:
       x1=x-f(x)/df(x)
       if abs(x-x1)<eps: break
       x=x1
    return x
def f(x):
    return 3*x**7-2*x**2+2*x-23
dx=0.0000001
zero=metoda_newtona(f, 1, 0.001, 0.000001)
print(zero)
```

```python
""" PY28 Lepszy szyfrant"""
class TSzyfrant:
    def __init__(self, klucz):
        self.klucz=klucz
    def szyfruj(self, liczba):
        return (liczba + self.klucz) % 256
    def odszyfruj(self, liczba):
        return (256 + liczba - self.klucz) % 256
    
class TSolitaire(TSzyfrant):
    def s_szyfruj(self, napis):
        wynik=[]
        for znak in napis:
            wynik.append(self.szyfruj(ord(znak)))
        return wynik

    def s_odszyfruj(self, zakodowany):
        wynik=''
        for kod in zakodowany:
            wynik= wynik+chr(self.odszyfruj(kod))
        return wynik

szyfrant=TSolitaire(1342)
wiadomosc = input('Podaj dodatnią liczbę do zaszyfrowania: ')
szyfr=szyfrant.s_szyfruj(wiadomosc)
print('Wynik=%s' % szyfr)
print('Odszyfrowany: %s' % szyfrant.s_odszyfruj(szyfr))
```

```python
""" PY30 Przyklady dziedziczenia"""
class SchoolMember:
    u'''Reprezentuje człowieka związanego z uczelnią.'''
    def __init__(self, imie, wiek):
        self.imie = imie
        self.wiek = wiek
        print('(Inicjalizacja SchoolMember: %s)' % self.imie)

    def powiedz(self):
        u'''Opowiedz o sobie.'''
        print('Imię:"%s" Wiek:"%s"' % (self.imie, self.wiek),)

class Wykladowca(SchoolMember):
    u'''Reprezentuje wykładowcę.'''
    def __init__(self, imie, wiek, pensja):
        SchoolMember.__init__(self, imie, wiek)
        self.pensja = pensja
        print('(Inicjalizacja Wykladowcy: %s)' % self.imie)
    def powiedz(self):
        SchoolMember.powiedz(self)
        print('Pensja: "%d"' % self.pensja)
class Student(SchoolMember):
    '''Reprezentuje studenta.'''
    def __init__(self, imie, wiek, oceny):
        SchoolMember.__init__(self, imie, wiek)
        self.oceny = oceny
        print('(Inicjalizacja Studenta: %s)' % self.imie)
    def powiedz(self):
        SchoolMember.powiedz(self)
        print('Oceny: "%d"' % self.oceny)

w = Wykladowca('Mrs. Shrividya', 40, 30000)
s = Student('Swaroop', 25, 75)

print() # wypisuje pustą linię
osoby = [w, s]
for osoba in osoby:
    osoba.powiedz() # działa zarówno dla Wykładowców, jak i Studentów
```

```python
""" PY32 Operacje na macierzach"""
class Macierz:
    def __init__(self, *x):
        self.x=x        
    def __repr__(self):
        return str(self.x)
    def __add__(self, inna):
        return Macierz(*([self.x[i]+inna.x[i] for i in range(4)]))        
    def __sub__(self, inna):
        return Macierz(*([self.x[i]-inna.x[i] for i in range(4)]))
    def __neg__(self):
        return Macierz(*([-self.x[i] for i in range(4)]))
    def __sub__(self, inna):
        return Macierz(*([self.x[i]-inna.x[i] for i in range(4)])) 
    def __mul__(self, inna):
        u=[self.x[0]*inna.x[0]+self.x[1]*inna.x[2], self.x[0]*inna.x[1]+self.x[1]*inna.x[3], \
           self.x[2]*inna.x[0]+self.x[2]*inna.x[2], self.x[2]*inna.x[1]+self.x[3]*inna.x[3]]
        return Macierz(*u)
    def det(self):
        return self.x[0]*self.x[3]-self.x[1]*self.x[2]
    
a=Macierz(*[1,2,3,4])
b=Macierz(*[3,1,0,6])
c=a-b
print(c.__repr__)
c=a*b
print(c.__repr__)
print(c.det)
```

```python
"""Przykład PY43. Synchronizacja wątków."""
import threading
import time

class mythread(threading.Thread): # Klasa mythread dziedziczącą z Thread 
    def __init__(self, threadID, name, counter):
#Musimy nadpisać metodę __init__ jednak najpierw! wywołując Thread.__init__
            threading.Thread.__init__(self)
            self.threadID=threadID
            self.name=name
            self.counter=counter
    def run(self):
            # Powinniśmy też nadpisać metodę run w której powinniśmy podać 
            # co należy wykonać w ramach danego wątku (W tym przykładzie 
            # nie robimy praktycznie nic (wywołujemy procedurę print(time)
            print("Rozpoczynamy wątek " + self.name +"\n")
            #threadLock.acquire()
            printS(time, self.name, self.counter, 5)
            #threadLock.release()
def printS(time, threadName, delay, counter):
    while counter:
            time.sleep(delay)
            print(" %s: %s" % (threadName, time.ctime(time.time())))
            counter-=1
#threadLock=threading.Lock()
threads=[]

# Tworzymy dwa wątki
thread1=mythread(1, "Nr 1", 1)
thread2=mythread(2, "Nr 2", 2)

# Uruchamiamy oba wątki (metoda start uruchamia run)
thread1.start()
thread2.start()

# Dodajemy oba wątki do krotki threads
threads.append(thread1)
threads.append(thread2)

# Czekamy póki wątki aż sie nie skończą
for t in threads:
      t.join()
print("Koniec")
```
