```python
"""Przykład PY02. Etykiety i przyciski."""
from tkinter import *

class LabelDemo( Frame ):
    """Dalsza ilustracja etykiet i przyciski"""
    def __init__( self ):
         """Tworzy trzy etykiety i dwa przyciski oraz pakuje je"""
         # tworzy obiekt ramka w którym umieszczam widgety
         # ramka jest upakowana w głównym oknie i ma tytuł - Przykład
         Frame.__init__( self ) 
         self.pack( expand = YES, fill = BOTH )
         self.master.title( "Przyklad" )

         # Tworzymy dwa przyciski o nazwach QUIT i Hej których wciśnięcie 
         # powoduje wywołanie metod quit1 i say_hi odpofrom Tkinter import *
         # Nazwy obiektów to button i hi_there 
         self.button = Button(self, text="QUIT", fg="red", 
               command=self.quit1, width=16, height = 1)
         self.hi_there = Button(self, text="Hej", command=self.say_hi, 
               width=16, height = 1)
         # Tworzymy trzy pola etykiet – dwa tekstowe i ikonę
         self.Label1 = Label( self, text = "Etykieta tekstowa" )
         self.Label2 = Label( self,text = "Etykieta tekstowa z ikona" )    
         self.Label3 = Label( self, bitmap = "warning" )
         # Teraz pakujemy wszystko w ramce
         self.button.pack(side=BOTTOM)
         self.hi_there.pack(side=BOTTOM)
         self.Label1.pack()
         self.Label2.pack( side = LEFT)
         self.Label3.pack( side = LEFT )
            
    # Metoda wywoływana przez wciśnięcie przycisku QUIT
    def quit1(self):
         print("Koniec")
         quit()         
            
    # Metoda wywpływana przez wciśnięcie Hej
    def say_hi(self):
         print("Hej - witojcie!")
# (A)Program główny –utworzenie okna LabelDemo i obsługa wszystkich zdarzeń            
def main():
    LabelDemo().mainloop()
# Jeśli nazwa uruchamianego programu to __main__ wywołaj procedurę main(). 
if __name__ == "__main__":
    main()
```

## <a name="et2"></a> 2. Widget Entry

```python
"""Przykład PY03. Różne pola tekstowe."""
from tkinter import *
from tkinter import messagebox
    
class EntryDemo( Frame ):
    """ Demonstruje cztery pola tekstowe"""
    def __init__( self ):
         """ Tworzy, pakuje i binduje zdarzenia do pól tekstowych"""
         Frame.__init__( self )
         self.pack( expand = YES, fill = BOTH )
         self.master.title( " Testowanie pól tekstowych" )
         self.master.geometry( "325x100" ) # width x length

         self.frame1 = Frame( self )
         self.frame1.pack( pady = 5 )

         self.text1 = Entry( self.frame1, name = "tekst 1" )
         self.text1.bind( "<Return>", self.showContents )
         self.text1.pack( side = LEFT, padx = 5 )

         self.text2 = Entry( self.frame1, name = "tekst 2" )
         self.text2.insert( INSERT, " Wpisz tekst tutaj" )
         self.text2.bind( "<Return>", self.showContents )
         self.text2.pack( side = LEFT, padx = 5 )

         self.frame2 = Frame( self )
         self.frame2.pack( pady = 5 )

         self.text3 = Entry( self.frame2, name = "tekst 3" )
         self.text3.insert( INSERT, " Pole nieedytowalne" )
         self.text3.config( state = DISABLED )
         self.text3.bind( "<Return>", self.showContents )
         self.text3.pack( side = LEFT, padx = 5 )

         # tekst in polu text4 pojawia się jako *
         self.text4 = Entry( self.frame2, name = "tekst 4", show = "*" )
         self.text4.insert( INSERT, "Tekst ukryty" )
         self.text4.bind( "<Return>", self.showContents )
         self.text4.pack( side = LEFT, padx = 5 )

    def showContents( self, event ):
         """Wyswietl zawartość pola"""
         # wyświetl nazwę pola tekstowego generującego zdarzenie
         theName = event.widget.winfo_name()
         # podstaw zawartość tego pola do zmiennej theContents
         theContents = event.widget.get()
         # Po wypełnieniu pola tekstowego i wciśnieciu ENTER 
         # Zawartość tego pola znajduje się w theContents
         # a tu dla celów szkoleniowych drukujemy je w osobnym
         # oknie modułu MessageBox
         messagebox.showinfo( "Komunikat", theName + ": " + theContents )
def main():
    EntryDemo().mainloop()
if __name__ == "__main__":
    main()
```

#### Niektóre metody geometrii grid:
* w.grid_configure(option=value, ...) - Umieść widget w siatce grid opisanej opcjami.
* w.grid_forget() - Usuń widget z siatki (może on być dalej przetwarzany)

* w.grid_remove() - Tak jak grid_forget()

```python
"""Przykład PY04. Program realizujący projekt widgetu"""
import tkinter
import image
# import ImageTk
 
class Example_grid(object):
    def __init__(self, root):                      # Definiujemy ramkę 
           self.mframe=tkinter.Frame(root)
           # Wczytanie danych do programu - obrazu
           global photo                          
           photo = PhotoImage(file='Data/PYTHONI/images.gif')
           # Definiujemy wszystkie potrzebne widgety            
           self.label1=tkinter.Label(self.mframe, text="Pierwszy ")
           self.label2=tkinter.Label(self.mframe, text="Drugi    ")
           self.label3 = tkinter.Label( self.mframe, image=photo )           
           self.button1=tkinter.Button(self.mframe, text=u"Powi\u0119ksz", 
                 fg="red",command=None, width=16, height = 1)
           self.button2=tkinter.Button(self.mframe, text="Zmniejsz", 
                 fg="red", command=None, width=16, height = 1)
           var = 1
           self.chbut = tkinter.Checkbutton(self.mframe, 
                 text=u"Rozwi\u0144",variable=var)
           self.e1 = tkinter.Entry(self.mframe)
           self.e2 = tkinter.Entry(self.mframe)
           # Rozmieszczamy korzystając z geometrii grid
           self.label1.grid(row=0, sticky='W')
           self.label2.grid(row=1, sticky='W')
           self.label3.grid(row=0, column=2, columnspan=2, 
                 rowspan=2,sticky='WENS', padx=5, pady=5)            
           self.button1.grid(row=2, column=2)
           self.button2.grid(row=2, column=3)
           self.chbut.grid(columnspan=2, sticky='W') 
           self.e1.grid(row=0, column=1)
           self.e2.grid(row=1, column=1)

           # Umieszczamy ramkę w oknie   
           self.mframe.grid(in_  = root, column = 2, row = 1, ipady = 2,  
                 sticky = "news")
def main():
     root = tkinter.Tk()
     gui = Example_grid(root)
     root.title(u"Przyk\u0142ad geometrii grid")
     root.mainloop()
if __name__ == '__main__': main()
```

### <a name="et5"></a> 3.2 Geometria pack



```python
"""Przykład PY05. Przyciski w geometrii pack."""
import tkinter 
class App:
     def __init__(self, master):
         tkinter.Button(master, text='Przycisk 1').pack(side=tkinter.TOP)
         tkinter.Button(master, text='Przycisk 2').pack(side=tkinter.LEFT)
         tkinter.Button(master, text='Przycisk 3').pack(side=tkinter.RIGHT)
root = tkinter.Tk()
root.option_add('*font', ('verdana', 12, 'bold'))
root.title(u"Przyk\u0142ad geometrii pack ")
display = App(root)
root.mainloop()
```

W geometrii **pack**, traktując ramkę jako wolny obszar do wstawiania, staramy się po kolei wg. zamieszczonych opcji **side** wstawiać widgety. Opcje **padx, pady** to odstępy pomiędzy komórkami na widgety w poziomie i pionie, opcje **anchor** i **fill** związane są z tym, gdzie dany widget w komórce ma być umieszczony (jeśli jest jakieś wolne miejsce w komórce) a opcja **expand** kontroluje, czy widget będzie się rozszerzał, gdy użytkownik będzie całe okno powiększał.

## <a name="et7"></a> 4. Fonty

1. Deskryptory fontów: są to krotki np. **("Times", 10, "bold"), ("Helvetica", 10, "bold italic"), ("Symbol", 8)**, a jeśli nazwa fontu nie ma znaku spacji to mogą te elementy wystąpić w jednym łańcuchu np. **"Times 10 bold", "Helvetica 10 bold italic", "Symbol 8"**
2. Zdefiniowane przez użytkownika nazwy fontów: **Moduł tkFont** określa klasę **Font**, dla której można utworzyć instancję, np. **tkFont.Font(family=”Tahoma”,size=10,weight=tkFont.BOLD, slant=tkFont.ITALIC)**
   * Atrybuty klasy fontów są następujące:
         * family typu string. Nazwa rodziny fontów..
         * size typu integer. Rozmiar fontu w punktach, jeśli size>0 i w pixelach jeśli ujemne.
         * weight grubość fontu. Używaj NORMAL(domyślne) lub BOLD.
         * slant używaj NORMAL(domyślne) lub ITALIC.
         * underline wartość logiczna jeśli 1 (True) to podkreślony, domyślny 0.
         * overstrike wartość logiczna gdy 1 (True) to przekreślony, domyślny 0.
         * F.actual(\[font_option\])** Zwraca słownik ze wszystkimi ustawieniami fontu F. Jeżeli występuje font_option to zwraca tylko wartość ustawionej opcji.
         * list(tkFont . families())

```python
""" Przykład PY11 - Wyświetlanie slajdów """
import PIL
import tkinter 
from PIL import Image, ImageTk
from itertools import cycle

class App(Tk):
    def __init__(self, image_files, x, y, delay):
        Tk.__init__(self)        
        self.geometry('+{}+{}'.format(x,y))
        self.delay = delay
        #self.pictures = cycle((ImageTk.PhotoImage(file=image), image) for image in image_files)
        self.pictures = cycle(image for image in image_files)        
        self.pictures = self.pictures
        self.picture_display = Label(self)
        self.picture_display.pack()
        self.images = [] # to keep references to images.

    def show_slides(self):        
        img_name = next(self.pictures)
        image_pil = Image.open(img_name).resize((500, 500)) #<-- resize images here
        self.images.append(ImageTk.PhotoImage(image_pil))      
        self.picture_display.config(image=self.images[-1])
        self.title(img_name)
        self.after(self.delay, self.show_slides)

    def run(self):
        self.mainloop()

delay = 3500

image_files=['Data/PYTHONI/img0.gif','Data/PYTHONI/img1.gif','Data/PYTHONI/img2.gif','Data/PYTHONI/img3.gif',
        'Data/PYTHONI/img4.gif','Data/PYTHONI/img5.gif', 'Data/PYTHONI/img6.gif', 
        'Data/PYTHONI/img7.gif','Data/PYTHONI/img8.gif','Data/PYTHONI/img9.gif']

x = 200
y = 150
app = App(image_files,x,y,delay)
app.show_slides()
app.run()
```

## <a name="et11"></a> 8. Widget Frame

Widget **Frame** to kontener na inne widgety. Jest bardzo ważny do procesu organizowania i grupowania
innych widgetów. Można w nim zastosować siatkę (geometria grid)

```python
"""Przykład PY13. Przykład zastosowania widgetu frame."""
from tkinter import *
root=Tk()
root.title(u"Przyk\u0142ad ramek, relief\u00f3w i borderwidth")
of = [None]*5
for bdw in range(5):
    of[bdw] = Frame(root, borderwidth=0)
    Label(of[bdw], text='borderwidth = %d ' % bdw).pack(side=LEFT)
    ifx = 0
    iff = []
    for relief in [RAISED, SUNKEN, FLAT, RIDGE, GROOVE, SOLID]:
        iff.append(Frame(of[bdw], borderwidth=bdw, relief=relief))
        Label(iff[ifx], text=relief, width=10).pack(side=LEFT)
        iff[ifx].pack(side=LEFT, padx=7-bdw, pady=5+bdw)
        ifx = ifx+1
    of[bdw].pack()
f = Frame(root, width=300, height=110)
xf = Frame(f, relief=GROOVE, borderwidth=2)
Label(xf, text=u"Wci\u015bnij go!").pack(pady=10)
Button(xf, text="Pierwszy!", state=DISABLED).pack(side=LEFT, padx=5, pady=8)
Button(xf, text="Drugi!", command=root.quit).pack(side=RIGHT, padx=5, pady=8)
xf.place(relx=0.01, rely=0.125, anchor=NW)
Label(f, text='Ta ramka ma tekst').place(relx=.06, rely=0.125, anchor=W)
f.pack()

root.geometry("700x300")
root.mainloop()
```

```python
"""Przykład PY22. Wykres utworzony w widgecie Canvas."""
from tkinter import *

# Dane w formie par (x,y)
data=[(7, 78), (12, 63), (19, 51), (27, 67), (35, 110), (48,130), 
 (59,154), (75, 160), (83,185), (98, 223), (120,212), (160,170), (193,185)]
# dane najmniejsza wartość, największa i co ile ma byc podziałka
xmin, xmax, xh =0, 200, 20
ymin, ymax, yh =0, 250, 50

def zamien(x,y):	# Ta procedura zamienia współrzędne (x,y) na piksele
    return 100 + 300*(x-xmin)/(xmax-xmin), 250 - 200*(y-ymin)/(ymax-ymin)

ix=int((xmax-xmin)/xh+1)   # Ilość podziałek na osi OX
iy=int((ymax-ymin)/yh+1)   # Ilość podziałek na osi OY

root = Tk()
root.title(u'Przyk\u0142ad - Wykres punktowy')

canvas = Canvas(root, width=450, height=300, bg = 'white')
canvas.pack()
Button(root, text=u'Wyj\u015bcie', command=root.quit).pack()

canvas.create_line(100,250,400,250, width=2)
canvas.create_line(100,250,100,50,  width=2)

for i in range(ix):
    x,y = zamien (xmin+i*xh,ymin)
    canvas.create_line(x,250,x,245, width=2)
    canvas.create_text(x,254, text='%d'% (xmin+i*xh), anchor=N)

for i in range(iy):
    x,y=zamien(xmin, ymin+i*yh)
    canvas.create_line(100,y,105,y, width=2)
    canvas.create_text(96,y, text='%5.1f'% (ymin+i*yh), anchor=E)

scaled = []
for x,y in data:
    scaled.append(zamien(x,y))

canvas.create_line(scaled, fill='red')

for x,y in scaled:
    canvas.create_oval(x-6,y-6,x+6,y+6,width=1, outline='black', fill='SkyBlue1')
root.mainloop()
```

```python
"""Przykład PY24. Kontrolowanie rozmiaru kuli - widget Scale."""
from tkinter import *

class ScaleDemo( Frame ):
    """Przykład uzycia Scale"""
    def __init__( self ):
        Frame.__init__( self )
        self.pack( expand = YES, fill = BOTH )
        self.master.title( u"Przyk\u0142ad wid\u017cetu Scale" )
        self.master.geometry( "220x270" )
        self.control = Scale( self, from_ = 0, to = 200,
        orient = HORIZONTAL, command = self.updateCircle)
        self.control.pack( side = BOTTOM, fill = X )
        self.control.set( 10 )
        # Tworzy Canvas I rysuje kulę
        self.display = Canvas( self, bg = "white" )
        self.display.pack( expand = YES, fill = BOTH )

    def updateCircle( self, scaleValue ):
        """Usuwa koło, wyznacza nowy rozmiar i rysuje od nowa"""
        end = int( scaleValue ) + 10
        self.display.delete( "circle" )
        self.display.create_oval( 10, 10, end, end,
        fill = "green", tags = "circle" )

def main():
    ScaleDemo().mainloop()
if __name__ == "__main__":
        main()
```

```python
"""Przykład PY25. Rejestracja wszystkich możliwych zdarzeń."""    # Coś złego w zakończeniu
from tkinter import *
import Pmw

# Słownik eventdict zawiera prawie wszystkie mozliwe zdarzenia widgetu
eventDict = {'2': 'KeyPress', '3': 'KeyRelease', '4': 'ButtonPress', '5': 'ButtonRelease', '6': 'Motion', '7': 'Enter','8': 'Leave', '9': 'FocusIn', '10': 'FocusOut','12': 'Expose', '15': 'Visibility', '17': 'Destroy','18': 'Unmap', '19': 'Map', '21': 'Reparent', '22': 'Configure', '24': 'Gravity', '26': 'Circulate','28': 'Property', '32': 'Colormap','36': 'Activate', '37': 'Deactivate',}
root = Tk()

# Funkcja reakcji na dowolne zdarzenie widgetu. Reakcja ta ma charakter 
# zapisania informacji o zdarzeniu do przewijalnewgo pola 
# tekstowego, ale oczywiście łatwo ją przerobić na inne reakcje
def reportEvent(event):
    # W zmiennej rpt znajduje sie pełny opis zdarzenia 
    rpt = '\n\n%s' % (80*'=')
    rpt += '\nZdarzenie: typ=%s (%s)' % (event.type,eventDict.get(event.type, 'Nieznane'))
    rpt += '\ntime=%s' % event.time
    rpt += ' widget=%s' % event.widget
    rpt += ' x=%s, y=%s'% (event.x, event.y)
    rpt += ' x_root=%s, y_root=%s' % (event.x_root, event.y_root)
    rpt += '\nserial=%s' % event.serial
    rpt += ' num=%s' % event.num
    rpt += ' height=%s' % event.height
    rpt += ' width=%s' % event.width
    rpt += ' keysym=%s' % event.keysym
    rpt += ' ksNum=%s' % event.keysym_num
    try:
        rpt += ' focus=%s' % event.focus
    except:
        try:
            rpt += ' send=%s' % event.send_event
        except:
            pass
    # Przesłanie informacji o zdarzeniu do pola tekstowego
    text2.yview(END)
    text2.insert(END, rpt)

frame = Frame(root, takefocus=1, highlightthickness=2, width=400, height=200)
text = Entry(frame, width=20, takefocus=1, highlightthickness=2, font="Verdana 11")
text2 = Pmw.ScrolledText(frame, text_width=100, text_height=30, text_font=('arial', 
12, 'bold'))
for event in eventDict.values():
    frame.bind('<%s>' % event, reportEvent)
    text.bind('<%s>' % event, reportEvent)
text.pack()
text2.pack(fill=BOTH, expand=YES)
frame.pack()
text.focus_set()
root.mainloop()
```

Na zakończenie przykład jeszcze innej przeglądarki do zdjęć:

```python
""" Przyklad PY26 - Jeszcze inna przeglądarka do zdjęć """
from tkinter import *
from PIL import Image , ImageTk

class AplikacjaGUI (Frame , object ):
    def __init__ (self , master ):
        super (AplikacjaGUI,self).__init__( master )
        self.pack(fill =BOTH , expand = True)
        self.nrZdjecia =0
        self.wczytajZdjecia()
        self.stworzWidzety()
    def wczytajZdjecia(self):
        zdjecia =["Data/PYTHONI/img0.gif","Data/PYTHONI/img1.gif", "Data/PYTHONI/img2.gif", "Data/PYTHONI/img3.gif"]
        self.wczytaneZdjecia =[]
        for i in zdjecia :
            obraz = Image.open(i)
            self.wczytaneZdjecia.append(ImageTk.PhotoImage(obraz))
    def stworzWidzety(self):
        przyciskPoprzednie = Button(self ,text =" Poprzednie ", command = self.akcjaPoprzednie )
        przyciskPoprzednie.pack(side = LEFT )
        self.plotno = Canvas (self , width =400 , height =400)
        self.plotno.pack ( side = LEFT )
        self.plotno.create_image (200 ,200 , image = self.wczytaneZdjecia[self.nrZdjecia])
        przyciskNastepne = Button (self , text =" Nastepne ", command = self . akcjaNastepne )
        przyciskNastepne.pack(side = LEFT )
    def aktualizujWidzety ( self ):
        self.plotno.create_image (200 ,200 , image = self . wczytaneZdjecia [self.nrZdjecia])
    def akcjaPoprzednie(self):
        if self.nrZdjecia ==0:
            self.nrZdjecia =0
        elif self.nrZdjecia >0:
            self.nrZdjecia -=1
        self.aktualizujWidzety ()
    def akcjaNastepne (self):
        if self.nrZdjecia ==2:
            self.nrZdjecia =2
        elif self.nrZdjecia <2:
            self.nrZdjecia +=1
        self.aktualizujWidzety ()

glowneOkno = Tk ()
app = AplikacjaGUI ( glowneOkno )
glowneOkno . mainloop ()
```
