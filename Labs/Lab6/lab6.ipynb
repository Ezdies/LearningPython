```python
import math
pi = 3.14

x = [-pi+ 2 * pi/100*i for i in range(101)]
y = [math.cos(i) for i in x]
plt.plot(x, y)
```

```python
x = np.linspace(-pi, pi, 100)
plt.plot(x, np.cos(x), label = "$cos(x)$")
plt.legend()
plt.show()
```

```python
x = np.linspace(-pi, pi, 100)
plt.plot(x, np.exp(x), label = "$e^x$")
plt.legend()
plt.show()
```

```python
import random as rand

a = np.array([
    [[0.85, 0.04, 0], [-0.04, 0.85, 1.6]],
    [[-0.15, 0.28, 0], [0.26, 0.24, 0.44]],
    [[0.2, -0.26, 0], [0.23, 0.22, 1.6]],
    [[0, 0, 0], [0, 0.16, 0]]
])

x = [0]
y = [0]

n = 1000

for i in rand.choices([0, 1, 2, 3], weights=[85, 7, 7, 1], k=n):
    x.append(a[i][0][0]*x[-1] + a[i][0][1] * y[-1] + a[i][0][2])
    x.append(a[i][0][0]*x[-2] + a[i][1][1] * y[-1] + a[i][1][2])

plt.figure(figsize=(6, 10))
plt.plot(x, y, marker=".", linewidth=0, markerscale=0.3)
plt.show()
```
